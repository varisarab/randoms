```
# Necessary imports
!pip install transformers
from transformers import pipeline
```

## Sentiment Analysis

```
# Create pipeline instance
sentiment = pipeline('sentiment-analysis')

# Positive sentiment example
result = sentiment('I am really happy today')
print(result)
print()
# Negative sentiment example
result = sentiment('I am feeling super low on energy')
print(result)
```

## Question Answering

```
q_and_a = pipeline('question-answering')
result = q_and_a({
                   'question': 'What are we trying to do with this question answering?',
                   'context': 'We are testing NLP functionality.'
                 })
print(result)
print()
print(f"The answer to your question: {result['answer']}")
```

```
ner = pipeline('ner')
result = ner('Question answering is pretty cool')
print(result)
```

## Predicting Masks

```
mask = pipeline('fill-mask')
result = mask('Deep <mask> networks work superbly for images!')
print(result)
print()
for x in result:
    print(f"Predicted string: {x['token_str']}, score:{x['score']}")
```
