```python
from transformers import AutoModelWithLMHead, AutoTokenizer
import torch
tokenizer = AutoTokenizer.from_pretrained("microsoft/DialoGPT-small")
model = AutoModelWithLMHead.from_pretrained("microsoft/DialoGPT-small")
```

```python
for i in range(5):
    # encode the new input, add eos_token, return a tensor (Pytorch)
    new_input_ids = tokenizer.encode(input(">> User:") + tokenizer.eos_token, return_tensors='pt')
    # append the new user input tokens to the chat history
    chatbot_input_ids = torch.cat([history_ids, new_input_ids], dim=-1) if i > 0 else new_input_ids
    # generated a response while limiting the total chat history to 1000 tokens    
    history_ids = model.generate(
    chatbot_input_ids, max_length=1000,
    pad_token_id=tokenizer.eos_token_id
    )
# pretty print last ouput tokens from bot
    print("DialoGPT: {}".format(tokenizer.decode(history_ids[:, chatbot_input_ids.shape[-1]:][0], skip_special_tokens=True)))
```
